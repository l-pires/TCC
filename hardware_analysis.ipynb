```python
import matplotlib.pyplot as plt
```

```python
dist = [15, 14, 13, 12, 11, 10, 9, 8, 7, 6, 5, 4]
smartphone = [890, 881, 868, 852, 828, 801, 762, 717, 641, 522, 316, 174]
tablet = [941, 939, 935, 931, 925, 917, 906, 892, 866, 833, 786, 703]
espelho = [817, 801, 777, 759, 715, 660, 608, 530, 421, 295, 149, 122]
```

```python
plt.plot(dist, smartphone, '-o', label='Smartphone')
plt.plot(dist, tablet, '-o', label='Tablet')
plt.plot(dist, espelho, '-o', label='Caso ideal')

plt.xlabel('Distância (cm)')
plt.ylabel('Leitura do receptor')
plt.ylim([0, 1023])
plt.legend()
plt.show()
```
